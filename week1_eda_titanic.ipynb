## Task 04 — Missing Value Analysis

```python
# 1. Missing Values Count


print("===== Missing Values Per Column =====")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# 2. Missing Value Heatmap
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(df.isnull(), cbar=True, yticklabels=False) #cbar means Color Bar. It displays a legend on the right side.
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# 3. Separate Columns

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

categorical_cols = df.select_dtypes(
    include=['object', 'category', 'bool']
).columns

# 4. Impute Numeric Columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean()) # it will fill all missing numeric values with the mean of that column.

# 5. Impute Categorical Columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0]) # it will fill all missing categorical values with the most frequent values.

# 6. Verify Missing Values

print("\n===== Missing Values After Imputation =====")
print(df.isnull().sum())
```

## 1. What This Dataset Represents

The Titanic dataset comes from the passenger manifest of the RMS Titanic, which sank in the North Atlantic in April 1912 after hitting an iceberg. Out of roughly 2,224 people on board, more than 1,500 died, making it one of the worst maritime disasters in history. The version used here (loaded via Seaborn) has 891 passenger records and is a standard benchmark dataset for classification problems.

Each row represents one passenger. The columns describe their demographics, ticket and cabin details, family members travelling with them, and whether they survived.

The features fall into a few categories:
- **Numeric**: `age`, `fare`, `sibsp`, `parch`
- **Categorical**: `sex`, `embarked`, `class`, `who`, `deck`, `alone`
- **Target/binary**: `survived`, `adult_male`

## 2. Features Most Correlated with Survival

The correlation heatmap and the bar/count plots point to a few clear relationships.

**Positively linked to survival:**
- **Fare** (r ≈ 0.26). Higher fares are tied to First Class, and First Class passengers survived at much higher rates (Image 3, Image 5).
- **Parch** (r ≈ 0.08), a weak effect, but passengers with a small number of parents/children aboard had slightly better odds than those travelling alone.

**Negatively linked to survival:**
- **Pclass** (r ≈ −0.34), the strongest relationship in the heatmap. Since class 1 = First and class 3 = Third, this means survival dropped as class number went up. Image 3 makes this obvious: Third Class has by far the most non-survivors.
- **Age** (r ≈ −0.07), a very weak negative trend. Not strong enough to draw real conclusions from on its own.

**Strongest categorical predictor:**
- **Sex** doesn't show up in the numeric heatmap, but it's the most powerful signal in the whole dataset. Image 7 shows women survived at about 74%, men at about 19%. That's a much bigger gap than anything in the numeric features, and it matches the historical "women and children first" evacuation order.

## 3. Key Patterns and Anomalies

**Survival trends.** Survival wasn't evenly split, and it clearly wasn't random either — sex and class explain most of the variation.

**Passenger class.** First Class had more survivors than non-survivors, Second Class was close to even, and Third Class was overwhelmingly non-survivors (Image 3). This fits the historical detail that Third Class cabins were lower in the ship, further from the lifeboats.

**Gender.** The survival gap by sex (Image 7) is the single biggest pattern in the data, bigger than the class or fare effect.

**Fare.** First Class fares are spread out with a few extreme outliers above 500 (Image 5), while Second and Third Class fares sit in a tight low range. Image 4 shows higher-fare passengers tend to be survivors regardless of age, which lines up with the fare-survival correlation.

**Age.** The distribution (Image 2) is right-skewed, with most passengers in their 20s–30s, a smaller bump for young children, and a long tail out to the 70s and 80s. The sharp spike around age 30 is partly artificial, caused by filling missing ages with the column mean.

**Missing values.** The heatmap (Image 1) shows `deck` is missing for the majority of rows, `age` is missing for a meaningful chunk, and `embarked`/`embark_town` are missing for only a few rows. `deck` is missing too much to fill in reliably with mode imputation.

**Outliers.** A handful of First Class passengers paid fares above 500 (Images 4 and 5), well past the typical First Class fare (median around 60). These could skew a model if fare is used raw, without scaling.

## 4. Conclusion

Survival on the Titanic came down mainly to sex and class, with fare adding a secondary signal that overlaps heavily with class. Age, sibsp, and parch show weak individual correlations but might still help when combined with stronger features.

For modelling, the most useful features look like `sex`, `pclass`, and `fare`, with `age`, `sibsp`, and `parch` as supporting features. `deck` should probably be dropped or treated as its own "missing" category rather than imputed normally, since over half its values are absent.

Before building a model, a few things need fixing:
- Encode categorical columns (`sex`, `embarked`, `class`) numerically.
- Reconsider age imputation — mean imputation is fine for EDA but flattens the real spread. Imputing by group (e.g. median age per `pclass` and `sex`) would be more realistic.
- Scale or log-transform `fare` to deal with the skew and outliers.
- Drop or carefully handle `deck` given how much of it is missing.
- Watch for overlap between `fare` and `pclass`, since they encode similar information and including both raw might not add much.

The dataset is solid for a classification task, but the missing-data issues in `age` and `deck` need deliberate handling, not blanket imputation, if the model is going to generalise well.
